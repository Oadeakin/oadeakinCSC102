```python
import tkinter as tk
from tkinter import messagebox as msgbx

def request1(name,location,weight):
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x300")

    label_1 = tk.Label(window, text=f"Name: {name}\nLocation: {location}\nWeight: {weight}kg")   
    label_1.pack()
    label_2 = tk.Label(window, text="Amount: 10000")
    label_2.pack()
def request2(name,location,weight):
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x300")

    label_1 = tk.Label(window, text=f"Name: {name}\nLocation: {location}\nWeight: {weight}kg")   
    label_1.pack()
    label_2 = tk.Label(window, text="Amount: 10000")
    label_2.pack()
def request3(name,location,weight):
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x300")

    label_1 = tk.Label(window, text=f"Name: {name}\nLocation: {location}\nWeight: {weight}kg")   
    label_1.pack()
    label_2 = tk.Label(window, text="Amount: 3500")
    label_2.pack()
def request4(name,location,weight):
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x300")

    label_1 = tk.Label(window, text=f"Name: {name}\nLocation: {location}\nWeight: {weight}kg")   
    label_1.pack()
    label_2 = tk.Label(window, text="Amount: 5000")
    label_2.pack()
def submit():
    
    name = name_entry.get()
    location = location_entry.get()
    weight = float(weight_entry.get())

    if location == "Ibeju-Lekki Community" and weight >= 10: 
        request1(name,location,weight)
    elif location == "Epe" and weight >= 10:
        request2(name,location,weight)
    elif location == "Ibeju-Lekki Community" and weight < 10:
        request3(name,location,weight)
    elif location == "Epe" and weight < 10:
        request4(name,location,weight)
    else:
        msgbx.showerror("Invalid input")
        
root = tk.Tk()
root.title("Order Form")
root.geometry("500x200")

#Create location label and entry
name_label = tk.Label(root, text="Name")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

#Create location Label and entry
location_label = tk.Label(root, text="Location(Ibeju-Lekki Community/Epe)")
location_label.pack()
location_entry = tk.Entry(root)
location_entry.pack()

weight_label = tk.Label(root, text="Weight(kg)")
weight_label.pack()
weight_entry = tk.Entry(root)
weight_entry.pack()

#Create submit button
submit_button = tk.Button(root, text="Submit Purchase", command=submit)
submit_button.pack()

#Run the main event loop
root.mainloop()
```
